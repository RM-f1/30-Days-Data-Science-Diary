```python
import pandas as pd
```

```python
df=pd.read_csv("E-commerce Customer Behavior - Sheet1.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Age'],bins=10,kde=True,color='mediumpurple')
plt.title("DISTRIBUTION OF CUSTOMER AGE")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Gender',y='Items Purchased', data=df,palette='pastel')
plt.title('Items Purchased By Gender')
plt.show()
```

```python
sns.barplot(x='Membership Type',y='Total Spend',data=df,palette='Set2')
plt.title("TOTAL SPEND BY MEMBERSHIP TYPE")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='Satisfaction Level',data=df,palette='cool')
plt.title('Satisfaction Level Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Purples')
plt.title("Correlation Heatmap")
plt.show()
```

###  Summary:

- Most customers are aged between 2535 years.
- Female customers tend to purchase slightly more items on average.
- Premium and Gold members have the highest total spend.
- "High" satisfaction levels dominate the dataset.
- Positive correlation observed between Total Spend and Items Purchased.
